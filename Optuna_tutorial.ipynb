# Function to minimize: 4x<sup>10</sup>-8y<sup>7</sup>+6z<sup>6</sup>+8w<sup>5</sup>-7a<sup>3</sup>-3b<sup>2</sup>

```
def objective(trial):
  params = {
      'x': trial.suggest_int('x',1,1000),
      'y': trial.suggest_uniform('y',-5,5),
      'z': trial.suggest_int('z',1,1000),
      'w': trial.suggest_loguniform('w',1e-3,10),
      'a': trial.suggest_uniform('a',2,6),
      'b': trial.suggest_categorical('b',[2,8,10]),
      'c': trial.suggest_categorical('c',['divide_by_2','multiply_with_2']),
  }

  equation = 4*(params['x']**10)-8*(params['y']**7)+6*(params['z']**6)+8*(params['w']**5)-7*(params['a']**3)-3*(params['b']**2)

  if params['c'] == 'divide_by_2':
    equation = equation / 2
  else:
    equation = equation * 2
  
  return equation
```

```
study = optuna.create_study()
study.optimize(objective, n_trials=1000)
```

```
print(study.best_params)
```

```
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=1000)
```

```
def objective(trial):
  params = {
      'x': trial.suggest_int('x',1,1000),
      'y': trial.suggest_uniform('y',-5,5),
      'z': trial.suggest_int('z',1,1000),
      'w': trial.suggest_loguniform('w',1e-3,10),
      'a': trial.suggest_uniform('a',2,6),
      'b': trial.suggest_categorical('b',[2,8,10]),
      'c': trial.suggest_categorical('c',['divide_by_2','multiply_with_2']),
  }

  equation = 4*(params['x']**10)-8*(params['y']**7)+6*(params['z']**6)+8*(params['w']**5)-7*(params['a']**3)-3*(params['b']**2)
  trial.report(equation,1)
  if params['c'] == 'divide_by_2':
    equation = equation / 2
  else:
    equation = equation * 2
  trial.report(equation,1)
  # Handle pruning based on the intermediate value.
  if trial.should_prune():
      raise optuna.structs.TrialPruned()
  
  return equation
```

```
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=1000)
```
